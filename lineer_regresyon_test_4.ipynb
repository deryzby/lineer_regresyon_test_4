```python
import pandas as pd
```

```python
reno_clio_verisi = pd.read_csv("reno_clio.csv")
reno_clio_verisi
```

```python
reno_clio_verisi['yas'] = 2024 - reno_clio_verisi['yıl']
reno_clio_verisi
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(reno_clio_verisi['yas'], reno_clio_verisi['km'], color = 'red')
plt.xlabel("Arabanın Yaşı")
plt.ylabel("Arabanın KM Bilgisi")
plt.title("Reno-Clio Arabanın Yaşı ve KM Bilgisi")
plt.show()
```

```python
# Lineer Regresyon modeli
# 1. formül y = ax + b
# 2. parametreler a = 3 b = 12
# olmak üzere 2 parçadan oluşur.
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(reno_clio_verisi[['yas']], reno_clio_verisi['km'])
```

```python
# modelki a ve b değerleri
model.coef_, model.intercept_
```

```python
tahmin = model.predict([[20]])
tahmin
```

```python
tahminler = model.predict(reno_clio_verisi[['yas']])
```
